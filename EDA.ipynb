# Average watch time VS Average quiz score

```python
plt.figure(figsize=(8,5))
sns.regplot(
    x='avg_watch_time',
    y='avg_quiz_score',
    data=df,
    scatter_kws={'alpha':0.2, 's':20},
    line_kws={'color':'red'}
)

plt.title("Average Watch Time vs Average Quiz Score")
plt.xlabel("Average Watch Time")
plt.ylabel("Average Quiz Score")
plt.show()
```

## Correaltion

```python
df[['avg_watch_time', 'avg_quiz_score']].corr()
```

## Observation
The correlation coefficient between average watch time and average quiz score is 0.11, indicating a very weak positive relationship. This suggests that students who watch more course content may achieve slightly higher quiz scores, but the relationship is not strong. Therefore, average watch time alone is not a reliable predictor of quiz performance, and other factors may also influence students' quiz scores.

# Completion rate vs Dropout rate

```python
plt.figure(figsize=(8,5))
sns.regplot(
    x='completion_rate',
    y='dropout_rate',
    data=df,
    scatter_kws={'alpha':0.2, 's':20},
    line_kws={'color':'red'}
)

plt.title("Completion Rate vs Dropout Rate")
plt.xlabel("Completion Rate")
plt.ylabel("Dropout Rate")
plt.show()
```

## Correaltion

```python
df[['completion_rate', 'dropout_rate']].corr()
```

## Observation
The correlation coefficient between completion rate and dropout rate is -0.95, indicating a very strong negative relationship. This means that courses with higher completion rates consistently have lower dropout rates. As the completion rate increases, the dropout rate decreases significantly, making this the strongest relationship observed in the dataset.

# Correaltion Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

## Observation
The correlation heatmap provides an overview of the relationships between all numerical variables in the dataset. The strongest negative correlation is observed between completion rate and dropout rate (-0.95). Most other variables exhibit weak positive or negative correlations, indicating that multiple factors collectively influence instructor effectiveness and student performance.

# Top Performing Instructors

```python
top_instructors = df.groupby('instructor_id').agg({
    'avg_feedback_score':'mean',
    'completion_rate':'mean',
    'avg_score_improvement':'mean'
}).sort_values(by='avg_feedback_score', ascending=False)

top_instructors.head(10)
```

## Observation
The table identifies the top 10 instructors based on their average feedback scores. These instructors also demonstrate strong completion rates and student score improvements, making them high-performing instructors in the dataset.

# Course wise analysis

```python
course_analysis = df.groupby('course_id').agg({
    'completion_rate':'mean',
    'dropout_rate':'mean',
    'avg_quiz_score':'mean'
}).sort_values(by='completion_rate', ascending=False)

course_analysis
```

## Observation
The course-wise analysis compares the average completion rate, dropout rate, and quiz score for each course. This helps identify courses with better student outcomes and highlights courses that may require improvements.




# Conclusion
The dataset contained 2,000 records with no missing values or duplicate rows, so no data cleaning was required.

Most performance metrics followed an approximately normal distribution.

Students generally had high assignment submission rates, watch time, and instructor feedback scores.

The relationship between average watch time and quiz scores was very weak (0.11), suggesting that watch time alone does not strongly influence quiz performance.

Assignment submission rate showed a weak positive relationship with completion rate (0.22).

Feedback response rate had a weak to moderate negative relationship with dropout rate (-0.27), suggesting that active instructor engagement may help improve student retention.

The strongest finding was the very strong negative correlation (-0.95) between completion rate and dropout rate, indicating that courses with higher completion rates consistently have lower dropout rates.
